> Consulte [este tutorial biblioteca `datetime`](https://www.w3schools.com/python/python_datetime.asp) caso n√£o esteja familiarizado.

```python
from datetime import datetime

class Evento:
    # Atributo de classe para contar o n√∫mero total de eventos
    total_eventos = 0  

    def __init__(self, titulo: str, data_hora: datetime, descricao: str):
        self.titulo = titulo
        self.data_hora = data_hora
        self.descricao = descricao
        self.is_concluido = False  # Evento come√ßa como n√£o conclu√≠do
        
        # Incrementa o total de eventos
        Evento.total_eventos += 1  

# Criando dois eventos para testar
evento1 = Evento("Reuni√£o de Trabalho", datetime(2025, 4, 10, 14, 30), "Discuss√£o sobre metas do trimestre.")
evento2 = Evento("Anivers√°rio do Jo√£o", datetime(2025, 5, 1, 19, 0), "Festa de anivers√°rio do Jo√£o no sal√£o.")

# Imprimindo os atributos de cada evento
print(f"Evento 1: {evento1.titulo}, {evento1.data_hora}, {evento1.descricao}, Conclu√≠do: {evento1.is_concluido}")
print(f"Evento 2: {evento2.titulo}, {evento2.data_hora}, {evento2.descricao}, Conclu√≠do: {evento2.is_concluido}")

# Imprimindo o total de eventos criados
print(f"Total de eventos criados: {Evento.total_eventos}")
```

#### Q2.
Agora vamos adicionar m√©todos √† nossa classe. Lembre dos diferentes decoradores que aprendemos. Voc√™ deve redefinir a classe com os seguintes m√©todos:
* Um m√©todo `isConcluido()` que avalia se a `data_hora` do evento √© menor que `datetime.now()` (a data e hora atual). Em caso positivo, atualiza o atributo de inst√¢ncia `is_concluido` para o valor `True`. 
* Um m√©todo de classe `num_eventos()` que retorna o valor do atributo de classe `total_eventos`.
* Um m√©todo est√°tico `valida_evento(nome, data_hora, descricao)` que recebe os atributos de um evento e testa os tipos de cada vari√°vel, retornando `True` caso todos estejam corretos e `False` caso contr√°rio. Para o teste, use a fun√ß√£o nativa do Python [`isinstance`](https://www.w3schools.com/python/ref_func_isinstance.asp).

Para testar sua classe atualizada:
* Crie uma inst√¢ncia de Evento com valor passado e chame o m√©todo `isConcluido()` para a inst√¢ncia criada. Em seguida imprima o atributo `is_concluido`. 
* Invoque o m√©todo de classe `num_eventos()` a partir da classe Evento (ou seja, sem criar nenhuma inst√¢ncia).
* Chame o m√©todo est√°tico `valida_evento()` a partir da classe Evento. Experimente passar valores corretos e incorretos. 


```python
from datetime import datetime

class Evento:
    # Atributo de classe para contar o n√∫mero total de eventos
    total_eventos = 0  

    def __init__(self, titulo: str, data_hora: datetime, descricao: str):
        self.titulo = titulo
        self.data_hora = data_hora
        self.descricao = descricao
        self.is_concluido = False  # Evento come√ßa como n√£o conclu√≠do
        
        # Incrementa o total de eventos
        Evento.total_eventos += 1  

    def isConcluido(self):
        """Verifica se o evento j√° ocorreu e marca como conclu√≠do."""
        if self.data_hora < datetime.now():
            self.is_concluido = True

    @classmethod
    def num_eventos(cls):
        """Retorna o n√∫mero total de eventos criados."""
        return cls.total_eventos

    @staticmethod
    def valida_evento(titulo, data_hora, descricao):
        """Valida se os tipos dos atributos de um evento s√£o corretos."""
        return isinstance(titulo, str) and isinstance(data_hora, datetime) and isinstance(descricao, str)

# Criando um evento com data passada para testar isConcluido()
evento_passado = Evento("Evento Antigo", datetime(2023, 4, 1, 10, 0), "Reuni√£o que j√° ocorreu.")

# Chamando o m√©todo isConcluido() e verificando se o status mudou
evento_passado.isConcluido()
print(f"Evento conclu√≠do? {evento_passado.is_concluido}")  # Deve imprimir True

# Testando o m√©todo de classe num_eventos()
print(f"Total de eventos criados: {Evento.num_eventos()}")  # Deve imprimir 1

# Testando o m√©todo est√°tico valida_evento()
print(Evento.valida_evento("Teste", datetime(2025, 5, 10, 14, 0), "Descri√ß√£o v√°lida"))  # True
print(Evento.valida_evento(123, "Data errada", 456))  # False
```

#### Q3.

Vamos incluir m√©todos m√°gicos! ‚ú®ü™ÑüîÆ

Redefina a classe incluindo":
* M√©todo `__str__` que imprime os atributos do evento na forma `"Evento: titulo, Data: data_hora, Descri√ß√£o: descricao, Concluido: is_concluido"`.
* Implemente os m√©todos de compara√ß√£o `__eq__`, `__ne__`, `__lt__`, `__le__`, `__gt__` e `__ge__` para comparar eventos baseados no atributo `data_hora`. Esses m√©todos devem comparar duas inst√¢ncias de Evento e retornar os resultados apropriados (`True` ou `False`).

Para testar, crie duas inst√¢ncias de Evento com datas diferentes. Imprima as inst√¢ncias com a fun√ß√£o `print()` e apresente o resultado das compara√ß√µes entre eventos (`==`, `!=`, `<`, `<=`, `>`, `>=`).

```python
from datetime import datetime

class Evento:
    # Atributo de classe para contar o n√∫mero total de eventos
    total_eventos = 0  

    def __init__(self, titulo: str, data_hora: datetime, descricao: str):
        self.titulo = titulo
        self.data_hora = data_hora
        self.descricao = descricao
        self.is_concluido = False  # Evento come√ßa como n√£o conclu√≠do
        
        # Incrementa o total de eventos
        Evento.total_eventos += 1  

    def isConcluido(self):
        """Verifica se o evento j√° ocorreu e marca como conclu√≠do."""
        if self.data_hora < datetime.now():
            self.is_concluido = True

    @classmethod
    def num_eventos(cls):
        """Retorna o n√∫mero total de eventos criados."""
        return cls.total_eventos

    @staticmethod
    def valida_evento(titulo, data_hora, descricao):
        """Valida se os tipos dos atributos de um evento s√£o corretos."""
        return isinstance(titulo, str) and isinstance(data_hora, datetime) and isinstance(descricao, str)

    def __str__(self):
        """Representa√ß√£o textual do evento."""
        return f"Evento: {self.titulo}, Data: {self.data_hora}, Descri√ß√£o: {self.descricao}, Conclu√≠do: {self.is_concluido}"

    # M√©todos de compara√ß√£o baseados na data_hora
    def __eq__(self, outro):
        """Verifica se dois eventos ocorrem na mesma data e hora."""
        return self.data_hora == outro.data_hora

    def __ne__(self, outro):
        """Verifica se dois eventos ocorrem em datas diferentes."""
        return self.data_hora != outro.data_hora

    def __lt__(self, outro):
        """Verifica se este evento ocorre antes do outro."""
        return self.data_hora < outro.data_hora

    def __le__(self, outro):
        """Verifica se este evento ocorre antes ou ao mesmo tempo que o outro."""
        return self.data_hora <= outro.data_hora

    def __gt__(self, outro):
        """Verifica se este evento ocorre depois do outro."""
        return self.data_hora > outro.data_hora

    def __ge__(self, outro):
        """Verifica se este evento ocorre depois ou ao mesmo tempo que o outro."""
        return self.data_hora >= outro.data_hora


# Criando dois eventos com datas diferentes
evento1 = Evento("Reuni√£o", datetime(2025, 5, 10, 14, 0), "Reuni√£o de planejamento")
evento2 = Evento("Confer√™ncia", datetime(2025, 6, 15, 9, 0), "Evento anual da empresa")

# Testando __str__
print(evento1)
print(evento2)

# Testando compara√ß√µes entre eventos
print("\nCompara√ß√µes:")
print("evento1 == evento2:", evento1 == evento2)  # False
print("evento1 != evento2:", evento1 != evento2)  # True
print("evento1 < evento2:", evento1 < evento2)    # True
print("evento1 <= evento2:", evento1 <= evento2)  # True
print("evento1 > evento2:", evento1 > evento2)    # False
print("evento1 >= evento2:", evento1 >= evento2)  # False
```
